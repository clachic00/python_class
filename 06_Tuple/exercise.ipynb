```python
class Add : 
    def add(self, tuple_value):
        total = 0
        for v in range(tuple_value):
            total += v
            print(total)
            
no_wrap_tuples = (1, 2, 3)

add_calculator = Add()
add_calculator.add(no_wrap_tuples)
```

```python
class Add:
    def add(self, tuple_value):
        total = 0
        for v in tuple_value:   # 튜플 요소를 하나씩 꺼냄
            total += v
        print(total)

no_wrap_tuples = (1, 2, 3)

add_calculator = Add()
add_calculator.add(no_wrap_tuples)
```

```python
# 리스트
location = ['서울특별시', '부산광역시', '인천광역시', '광주광역시', '울산광역시']
print('광주광역시' in location)                   # True
print(['광주광역시', '부산광역시'] in location)    # False
print(['서울특별시', '부산광역시'] in location)    # False
```
