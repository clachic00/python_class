```python
""" 
    2025-11-26 양창일

    1. 숫자 랜덤 3개 선언 및 저장부

    2. 시작 > 숫자 input 받고 결과 출력
    
    3. 비교 및 결과 값 반환
    
    4. 값 형식 체크
    
"""
```

```python
import ipywidgets as widgets
from IPython.display import display

class BaseBallPlay:
    def __init__(self):
        self.main_number = []
        self.first_num = 0
        self.second_num = 0
        self.third_num = 0

    def start(self, before):
        # 값은 한번만 뽑아야함
        if len(self.main_number) < 1 :
            self.pick_numbers()

        # 값 받기
        if self.first_num == 0 :
            self.first_num = input(before + "\n\t1. 첫번 째 숫자를 입력해 주세요. [1~9]")
            if not self.type_check(self.first_num) :
                self.start()
                
        if self.second_num == 0 :        
            self.second_num = input("2. 두번 째 숫자를 입력해 주세요. [1~9]")
            if not self.type_check(self.second_num) :
                self.start()
                
        if self.third_num == 0 :        
            self.third_num = input("3. 세번 째 숫자를 입력해 주세요. [1~9]")
            if not self.type_check(self.third_num) :
                self.start()
            
        # 수정사항 self.start 변수에 몇번째 수 진행중인지 넣어야함 
            
        result = self.check_baseball(self.first_num, self.second_num, self.second_num)
        
        if not result : 
            input("[ 3 스트라이크! 정답입니다. 축하드립니다. ]")
            return
        
        self.start(result)

    def pick_numbers(self): # 숫자 세개 랜덤 뽑기
        import random
        self.main_number = random.sample(range(1, 10), 3)
        # print("뽑힌 숫자:", self.main_number)
    
    def type_check(self, check_num) : # 입력 숫자 검증
        if str.isdigit(check_num) :
        #   print(check_num)
            if check_num == "0" or len(check_num)>1 :
                return False
            else :
                return True
        else :
            return False
    def check_baseball(self, first, second, third) :  # 베이스볼 체크
        mn = self.main_number
        strike = 0
        ball = 0
        # print(first,second,third)
        if int(first) in mn and mn[0] == int(first) :
            strike += 1
        elif int(first) in mn and mn[0] != int(first) :
            ball += 1
        if int(second) in mn and mn[1] == int(second) :
            strike += 1
        elif int(second) in mn and mn[1] != int(second) :
            ball += 1
        if int(third) in mn and mn[2] == int(third) :
            strike += 1
        elif int(third) in mn and mn[2] != int(third) :
            ball += 1
        
        if strike == 3 :
            return False
        
        result =" [" + str(strike) + " 스트라이크\n" + str(ball) + " 볼 ]"
        
        return result 
    
    

bp = BaseBallPlay()
bp.start("[ Welcome to Baseball Play! ] ")
```
