```python
class Person:
    national = 'korea'
    
    def greeting(self):
        return "Hello. This is Python"
class Learner:
    def greeting(self):
        return "Hello. I am Learner"
    def learn(self):
        return "I am learning Python"
    
class Student(Person, Learner):
    pass

student = Student()
print(student.greeting())
print(student.learn())
```
