```python
num = int(input('정수 입력 : '))
print('홀수' if num % 2 != 0 else '짝수')
```

```python
a = int(input('첫번째 정수 입력 : '))
b = int(input('두번째 정수 입력 : '))

big = a if a > b else b
small = a if a < b else b
print(big, small)
```
