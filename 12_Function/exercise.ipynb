```python
def information(name, color):
    print(name + '의 최애 색상', color, sep=" : ")

information("capybara","black")
information(color="white", name="lion")
```

```python
information(color="black", "copybara") #에러 발생
information(name="copybara", "black") #에러 발생
```

```python
def add_many(*args):
    result = 0
    
    for i in args:
        result += i
    return result

add_many(1,2,3,4,5,6,8)   # 튜플로 묶임
```

```python
def show_info(**kwargs):
    print(kwargs)

show_info(name="A", lang="B", age=12)   # 딕셔너리로 묶임
```

```python
def sum(a,b,c):
    return a+ b + c

numbers = [1, 2, 3]
sum(*numbers)
```

```python
add = lambda x, y : x+y
add(10,4)
```

```python
minus = lambda x, y : x-y
minus(10,4)
```

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x : x**2 , numbers)
print(list(squared_numbers))
```

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x : x % 2==0, numbers)
print(list(even_numbers))
```

```python
points = [(1, 2), (3, 1), (5, -1)]
sorted_points = sorted(points, key = lambda x : x[1])
print(sorted_points)
```
