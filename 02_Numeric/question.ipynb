### **문제 3**

다음 조건을 만족하는 프로그램을 작성하세요.

1. `num1`에 15를 저장하고, `num2`에 4를 저장합니다.
2. 두 수의 **나머지**와 **몫**을 각각 구하여 출력하세요.

```python
num1, num2 = 15, 4
print("몫 : " + str(num1 % num2) + "\n나머지 : " + str(num1 // num2)) 
```

### **문제 5**

다음 조건에 맞는 프로그램을 작성하세요.
