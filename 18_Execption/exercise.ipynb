```python
class MyError(Exception):
    pass
def say_nick(nick):
    if nick == '돼지':
        raise MyError()
    print(nick)

try:
    say_nick("곰")
    say_nick("돼지")
    
except MyError:
    print("허용되지 않는 별명입니다.")
```
