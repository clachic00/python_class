### **문제 3: `.log` 파일 읽기**

다람쥐는 프로그램 실행 로그를 확인하려고 합니다.

1. `system.log` 파일을 다음 내용으로 생성하세요.
    
    ```
    [INFO] 2024-01-01 12:00:00: 시스템 시작
    [WARN] 2024-01-01 12:05:00: 메모리 사용량 증가
    [ERROR] 2024-01-01 12:10:00: 프로그램 종료
    ```
    
2. 파일을 읽어와 `WARN` 또는 `ERROR`가 포함된 줄만 출력하세요.

````python
with open("new/fi/le/system.log", "w", encoding="utf-8") as log_write :
    log_write.write("[INFO] 2024-01-01 12:00:00: 시스템 시작\n[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n[ERROR] 2024-01-01 12:10:00: 프로그램 종료")
    log_write.close()
    
with open("new/fi/le/system.log", "r", encoding="utf-8") as log_read :
    text_list = log_read.readlines()
    
    for row in text_list:
        if "ERROR" in row or "WARN" in row:
            print(row)
    log_read.close()
````

````python
with open("new/fi/le/system.log", "w", encoding="utf-8") as log_write :
    log_write.write("[INFO] 2024-01-01 12:00:00: 시스템 시작\n[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n[ERROR] 2024-01-01 12:10:00: 프로그램 종료")
    log_write.close()
    
with open("new/fi/le/system.log", "r", encoding="utf-8") as log_read :
    text_list = log_read.readlines()
    
    result_list = list(filter(lambda x : "ERROR" in x or "WARN" in x, text_list))
    print(result_list[0]+"\n"+result_list[1])
    log_read.close()
````

### **문제 4: `.json` 파일 생성 및 읽기**

다람쥐는 좋아하는 음식 데이터를 저장하고 싶습니다.

1. `favorite_foods.json` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```json
    {
        "name": "다람쥐",
        "favorite_foods": ["도토리", "호두", "밤"]
    }
    ```
    
2. 파일을 읽어와 다람쥐의 좋아하는 음식 리스트를 출력하세요.

````python
import ast
json_text = '{"name": "다람쥐","favorite_foods": ["도토리", "호두", "밤"]}'
with open("new/fi/le/favorite_foods.json","w",encoding="utf-8") as json_write :
    json_write.write(json_text)
    json_write.close()
with open("new/fi/le/favorite_foods.json","r",encoding="utf-8") as json_read :
    read_dict = json_read.read()    
    casting_dict = ast.literal_eval(read_dict)
    print(casting_dict["favorite_foods"])
````

````python
import json
dict1 = {"name": "다람쥐","favorite_foods": ["도토리", "호두", "밤"]}
with open("new/fi/le/favorite_foods.json","w",encoding="utf-8") as json_write :
    json.dump(dict1, json_write, ensure_ascii=False, indent=4)
    json_write.close()
with open("new/fi/le/favorite_foods.json","r",encoding="utf-8") as json_read :
    dict2 = json.load(json_read)
    print(dict2["favorite_foods"])
    json_read.close()    
````

### **문제 5: `.html` 파일 수정**

다람쥐는 자신의 이력서를 작성 중입니다. `resume.html` 파일을 다음 내용으로 생성하고 수정하세요.

1. `resume.html` 파일 생성 내용:
    
    ```html
    <html>
        <head><title>다람쥐의 이력서</title></head>
        <body>
            <h1>다람쥐의 이력서</h1>
            <p>경력: 신입</p>
        </body>
    </html>
    ```
    
2. 수정 작업:
    - `<p>` 태그 안의 "신입"을 "1년 경력"으로 변경하세요.
3. 수정된 파일을 읽어와 전체 내용을 출력하세요.

````python
from bs4 import BeautifulSoup

html_text = '<html>'
html_text +='    <head><title>다람쥐의 이력서</title></head>'
html_text +='    <body>'
html_text +='        <h1>다람쥐의 이력서</h1>'
html_text +='        <p>경력: 신입</p>'
html_text +='    </body>'
html_text += '</html>'

with open("new/fi/le/resume.html","w",encoding="utf-8") as html_write :
    html_write.write(html_text)
    html_write.close()

with open("new/fi/le/resume.html","r",encoding="utf-8") as html_read :
    html = html_read.read()
    soup = BeautifulSoup(html, "html.parser")
    p = soup.find("p")
    p.string = "1년 경력"
    html_read.close()
    
with open("new/fi/le/resume.html","w",encoding="utf-8") as html_update :
    html_update.write(str(soup))
    html_update.close()
    
with open("new/fi/le/resume.html","r",encoding="utf-8") as html_read_again :
    print(html_read_again.read())
    html_read_again.close()
    
````

````python
from bs4 import BeautifulSoup
html_text = """
<html>
    <head><title>다람쥐의 이력서</title></head>
    <body>
        <h1>다람쥐의 이력서</h1>
        <p>경력: 신입</p>
    </body>
</html>
"""

with open("new/fi/le/resume.html","w",encoding="utf-8") as html_write :
    html_write.write(html_text)
    html_write.close()

with open("new/fi/le/resume.html","r",encoding="utf-8") as html_read :
    html = html_read.read()
    soup = BeautifulSoup(html, "html.parser")
    p = soup.find("p")
    p.string = "1년 경력"
    html_read.close()
    
with open("new/fi/le/resume.html","w",encoding="utf-8") as html_update :
    html_update.write(soup.prettify())
    html_update.close()
    
with open("new/fi/le/resume.html","r",encoding="utf-8") as html_read_again :
    print(html_read_again.read())
    html_read_again.close()
````
